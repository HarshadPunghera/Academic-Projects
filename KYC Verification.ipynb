```python
class ValPAN:
   
    def validatePAN(PAN):
       
       
        if PAN.isalnum() == False:
           
            print("Invalid PAN Number")
            print("PAN Number contains non alpha numeric characters")


        elif len(PAN) != 10:
           
            print("Invalid PAN Number")
            print("PAN Number length is incorrect")

       
        elif PAN[3] != 'P':
           
            print("Entered PAN is of corporate customer. Please route the form via Corporate Section")
       
        else:
           
            print("Enetered PAN is valid PAN")

class ValPassport:
   
    def validatePassport(PassportNo):
             
        if PassportNo.isalnum() == False:
           
            print("Invalid Passport Number")
            print("Passport Number contains non alpha numeric characters")

        elif len(PassportNo) != 8:
           
            print("Invalid Passport Number")
            print("Passport Number length is incorrect")

        elif PassportNo[0] != 'J':
           
            print("Invalid Passport Number")
            print("Passport Number starts with J")
       
        else:
           
            print("Enetered Passport Number is valid")


class ValAadhar:

    def validateAadhar(AadharNo):
       
        if AadharNo.isdigit() == False:
           
            print("Invalid Aadhar Number")
            print("Aadhar Number contains non numeric characters")

        elif len(AadharNo) != 12:
           
            print("Invalid Aadhar Number")
            print("Aadhar Number length is incorrect")

        elif AadharNo[0] == '0' or AadharNo[0] == '1' :
           
            print("Invalid Aadhar Number")
            print("Aadhar Number does not start 0 or 1")
       
        else:
           
            print("Enetered Aadhar Number is valid")


class ValKYC(ValPAN,ValPassport,ValAadhar):
   
   
    def ValidateKYC(self):
       
        KYCDoc = input("Please enter type of KYC : ")
   
        if KYCDoc == "PAN":
           
            PAN = input("Enter PAN No")
            ValPAN.validatePAN(PAN)
       
        elif KYCDoc == "Passport":
       
            PassportNo = input("Enter Passport Number : ")
            ValPassport.validatePassport(PassportNo)
       
        elif KYCDoc == "Aadhar":
           
            AadharNo = input("Enter Aadhar Number : ")
            ValAadhar.validateAadhar(AadharNo)
       
        else:
       
            print("Incorrect KYC Doc entered")

ValKYC.ValidateKYC('self')
```
